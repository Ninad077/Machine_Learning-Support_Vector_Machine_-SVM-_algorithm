```python
import numpy as np
import pandas as pd
```

```python
# Read the training & test data
data = pd.read_csv('Bank_Personal_Loan_Modelling.csv',index_col=0, header=0)
```

```python
data
```

```python
# Checking for null values
data.isnull().sum()
```

```python
# Check for special characters 

for i in data.columns:
    print(f"Unique value for coulmn: {i}\n\n{data[i].unique()}\n") 
```

```python
x= data.values[:,0:8]
y= data.values[:,7]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(x)
x= scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.3)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_KNN = KNeighborsClassifier(n_neighbors=int(np.sqrt(len(X_train))),metric='euclidean')
```

```python
model_KNN.fit(X_train,Y_train)
```

```python
Y_pred = model_KNN.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm = confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report:")
print(classification_report(Y_test,Y_pred))

acc = accuracy_score(Y_test,Y_pred)
print("Accuracy of the model:",acc)
```

```python
# Accuracy score with KNN was 67.4%
```

```python
# Let us try with Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression

# Create a model object
classifier = LogisticRegression()

# Train the model object
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test)
print(Y_pred)
print(list(zip(Y_test,Y_pred))) # Here we are using List with zip to show us the list of predictions that are 
                                # classified correctly (e.g. (1,1), (0,0)) 
                                # & those are wrongly classified (e.g. (1,0),(0,1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm = confusion_matrix(Y_test, Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test,Y_pred)
print("accuracy of the model:",acc)
```

```python
# Logistic Regression accuracy : 67.86%
```
